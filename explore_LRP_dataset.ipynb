```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
# hf = h5py.File("data/CellSeg_dataset/LateralRootPrimordia_processed_wide_boundary_non_precropped/train/Movie1_t00003_crop_gt.h5")
hf = h5py.File("data/CellSeg_dataset/LateralRootPrimordia_processed_wide_boundary_non_precropped/train/Movie1_t00009_crop_gt.h5")
```

```python
raw_img = np.array(hf["raw"], dtype=np.float)
hand_seg = np.array(hf["ins"], dtype=np.float)

foreground = np.array(hf["foreground"], dtype=np.float)
background = np.array(hf["background"], dtype=np.float)
boundary = np.array(hf["boundary"], dtype=np.float)

edge_foreground = np.array(hf["edge_foreground"], dtype=np.float)
edge_background = np.array(hf["edge_background"], dtype=np.float)
edge_boundary = np.array(hf["edge_boundary"], dtype=np.float)
```

```python
N=200

plt.figure()
plt.title("raw_img")
plt.imshow(raw_img[:,:,N])
plt.figure()
plt.title("hand_seg")
plt.imshow(hand_seg[:,:,N])

plt.figure()
plt.title("foreground")
plt.imshow(foreground[:,:,N])
plt.figure()
plt.title("boundary")
plt.imshow(boundary[:,:,N])
plt.figure()
plt.title("background")
plt.imshow(background[:,:,N])

plt.figure()
plt.title("edge_foreground")
plt.imshow(edge_foreground[:,:,N])
plt.figure()
plt.title("edge_boundary")
plt.imshow(edge_boundary[:,:,N])
plt.figure()
plt.title("edge_background")
plt.imshow(edge_background[:,:,N])
```
